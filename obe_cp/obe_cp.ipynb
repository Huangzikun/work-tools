```python
import os
import pandas as pd

old_dir = "/Users/huangzikun/Desktop/桂林学院/面向对象程序设计/期末材料/期末试卷-2023数据科学与大数据技术2班"
new_dir = "/Users/huangzikun/Desktop/桂林学院/面向对象程序设计/期末材料_new/2023数据科学与大数据技术2班《面向对象程序设计》期末试卷黄子坤47份"
```

```python
file_list = os.listdir(old_dir)
file_list
```

```python
df = pd.read_excel("/Users/huangzikun/Desktop/桂林学院/面向对象程序设计/期末材料/名单.xlsx", names=['student_id','student_name', "1", "2"])
df
```

```python
df = df.apply(lambda x: x.str.replace('\t', ''))
df = df[df['student_id'].str.match(r'^\d+$')]
df
```

```python
# 遍历 DataFrame
for index, row in df.iterrows():
    student_id = str(row['student_id'])
    student_name = row['student_name']
    student_path = os.path.join(new_dir, f"{student_id}{student_name}")
    os.makedirs(student_path, exist_ok=True)
```

```python
import re
import shutil

count = 0

for index, row in df.iterrows():
    student_id = str(row['student_id'])
    student_name = row['student_name']
    student_path = os.path.join(new_dir, f"{student_id}{student_name}")
    
    for file in file_list:
        if student_id in file:
            shutil.copy(os.path.join(old_dir, file), student_path)
            print(f"复制{student_id}")
            count = count + 1
            break
        elif student_name in file:
            shutil.copy(os.path.join(old_dir, file), student_path)
            print(f"复制{student_id}")
            count = count + 1
            break

print(f"复制成功{count}")
```
